# Train the Model

```python
from ultralytics import YOLO

# Load pretrained model
model = YOLO("yolov8n.pt")

# Train the model
model.train(data="/content/Weqaa-7/data.yaml", epochs=150, batch=16)
```

#Evaluate model

```python
# Evaluate model performance on the validation set
metrics = model.val()

# Export the model to ONNX format
model.export(format="onnx")
```

```python
print("map50-95(B)", metrics.box.map)   # map50-95(B)
print("map50(B)", metrics.box.map50)  # map50(B)
print("map75(B)", metrics.box.map75)  # map75(B)
print(metrics.box.maps)   # a list contains map50-95(B) of each category

# print('\n\n\n')
# print("map50-95(M)", metrics.seg.map)   # map50-95(M)
# print("map50(M)", metrics.seg.map50)  # map50(M)
# print("map75(M)", metrics.seg.map75)  # map75(M)
# print(metrics.seg.maps )
```
